```python
from sympy import Symbol, plot

x = Symbol('x', real = True)
f = x**3 - 50*x
g = -(x**4) + 88*x**2 - 241

#найдем точки пересечения f и g, для этого
#составим равенство: (f == g) :---> (g - f == 0 == z)
z = g - f
z
```

```python
import numpy
#найдем корни полинома z:
#z = -(x**4) - x**3 + 88*x**2 + 50*x - 241
coeff = [-1, -1, 88 ,50, -241]
x_roots = numpy.roots(coeff)
x_roots
```

```python
#найдем координаты точек пересечения:
j = 0
y_roots = []
for i in x_roots:
    y_roots.append(f.subs(x, i))
    print (f'x{(j:=j+1)}={i}     y{j}={y_roots[-1]}')
```

```python
#графики функций f и g
plot(f, g)
```

```python
#график функции пересечения z
plot(z,(x,-12, 12))
```
